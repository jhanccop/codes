## DETECCION DE FRAUDES

Indicadores comunes para detectar posible fraude bancario. El script analiza:

1. Transacciones de alto valor: Identifica transacciones que estÃ¡n en el top 5% del monto


2. MÃºltiples transacciones en corto tiempo: Detecta cuando hay 3 o mÃ¡s transacciones en una hora desde la misma cuenta


3. Patrones inusuales de ubicaciÃ³n: Alerta cuando una cuenta realiza transacciones en mÃºltiples ubicaciones en 24 horas


4. Transacciones en horarios inusuales: Identifica operaciones realizadas entre las 23:00 y 4:00



Para cada alerta, el sistema:

* Asigna un nivel de riesgo (Alto, Medio, Bajo)
* Registra la fecha y hora
* Proporciona una descripciÃ³n detallada
* Mantiene referencia a la transacciÃ³n original

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
def detectar_fraude(transacciones_df):
    df = transacciones_df.copy()
    df['fecha'] = pd.to_datetime(df['fecha'])
    alertas = []
    
    # 1. Transacciones de alto valor
    umbral_alto_valor = df['monto'].quantile(0.95)  # Top 5% de transacciones
    transacciones_grandes = df[df['monto'] >= umbral_alto_valor]
    for _, trans in transacciones_grandes.iterrows():
        alertas.append({
            'fecha': trans['fecha'],
            'tipo_alerta': 'TransacciÃ³n de alto valor',
            'descripcion': f'Monto: {trans["monto"]:,.2f}',
            'nivel_riesgo': 'Medio',
            'id_transaccion': trans.name
        })

    print("uno hecho")
    
    # 2. MÃºltiples transacciones en corto tiempo
    def detectar_transacciones_rapidas(grupo):
        if len(grupo) >= 3:  # MÃ¡s de 3 transacciones en 1 hora
            return grupo
        return pd.DataFrame()
    
    ventana_tiempo = df.sort_values('fecha').set_index('fecha')
    transacciones_rapidas = ventana_tiempo.groupby('cuenta_origen').rolling("1h")#.apply(lambda x: len(x) >= 3).reset_index()
    print(transacciones_rapidas)
    for cuenta, grupo in df.groupby('cuenta_origen'):
        if len(grupo) >= 3:
            tiempo_min = grupo['fecha'].min()
            tiempo_max = grupo['fecha'].max()
            if (tiempo_max - tiempo_min) <= timedelta(hours=1):
                alertas.append({
                    'fecha': tiempo_max,
                    'tipo_alerta': 'MÃºltiples transacciones rÃ¡pidas',
                    'descripcion': f'Cuenta {cuenta}: {len(grupo)} transacciones en {(tiempo_max - tiempo_min).seconds/3600:.1f} horas',
                    'nivel_riesgo': 'Alto',
                    'id_transaccion': grupo.index[0]
                })

    print("DOS hecho")
    
    # 3. Patrones inusuales de ubicaciÃ³n
    def detectar_ubicaciones_sospechosas(grupo):
        if len(grupo['ubicacion'].unique()) > 2:  # MÃ¡s de 2 ubicaciones diferentes en 24h
            return grupo
        return pd.DataFrame()
    
    ubicaciones_sospechosas = df.groupby('cuenta_origen').apply(detectar_ubicaciones_sospechosas)
    for _, grupo in ubicaciones_sospechosas.groupby('cuenta_origen'):
        alertas.append({
            'fecha': grupo['fecha'].max(),
            'tipo_alerta': 'MÃºltiples ubicaciones',
            'descripcion': f'Transacciones en {len(grupo["ubicacion"].unique())} ubicaciones diferentes en 24h',
            'nivel_riesgo': 'Alto',
            'id_transaccion': grupo.index[0]
        })
    
    # 4. Transacciones en horarios inusuales
    horario_inusual = df[
        (df['fecha'].dt.hour >= 23) | 
        (df['fecha'].dt.hour <= 4)
    ]
    for _, trans in horario_inusual.iterrows():
        alertas.append({
            'fecha': trans['fecha'],
            'tipo_alerta': 'Horario inusual',
            'descripcion': f'TransacciÃ³n a las {trans["fecha"].strftime("%H:%M")}',
            'nivel_riesgo': 'Bajo',
            'id_transaccion': trans.name
        })
    
    # Crear DataFrame de alertas
    alertas_df = pd.DataFrame(alertas)
    
    # Resumen de indicadores
    resumen = {
        'total_transacciones': len(df),
        'total_alertas': len(alertas),
        'alertas_alto_riesgo': len(alertas_df[alertas_df['nivel_riesgo'] == 'Alto']),
        'monto_total_alertas': df.loc[alertas_df['id_transaccion']]['monto'].sum(),
        'cuentas_afectadas': len(alertas_df['id_transaccion'].unique())
    }
    
    return alertas_df, resumen
```

```python
if __name__ == "__main__":
    # Crear datos de ejemplo
    data = {
        'fecha': pd.date_range(start='2024-01-01', periods=100, freq='h'),
        'monto': np.random.normal(1000, 500, 100),
        'tipo_transaccion': np.random.choice(['transferencia', 'retiro', 'deposito'], 100),
        'ubicacion': np.random.choice(['Madrid', 'Barcelona', 'Valencia', 'Sevilla'], 100),
        'cuenta_origen': np.random.choice(['cuenta1', 'cuenta2', 'cuenta3'], 100),
        'cuenta_destino': np.random.choice(['destino1', 'destino2', 'destino3'], 100)
    }
    
    df = pd.DataFrame(data)
    alertas, resumen = detectar_fraude(df)
    print("\nResumen de alertas detectadas:")
    print(f"Total de transacciones analizadas: {resumen['total_transacciones']}")
    print(f"Total de alertas generadas: {resumen['total_alertas']}")
    print(f"Alertas de alto riesgo: {resumen['alertas_alto_riesgo']}")
```
