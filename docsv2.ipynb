## DETECCION DE FRAUDES

Indicadores comunes para detectar posible fraude bancario. El script analiza:

1. Transacciones de alto valor: Identifica transacciones que están en el top 5% del monto


2. Múltiples transacciones en corto tiempo: Detecta cuando hay 3 o más transacciones en una hora desde la misma cuenta


3. Patrones inusuales de ubicación: Alerta cuando una cuenta realiza transacciones en múltiples ubicaciones en 24 horas


4. Transacciones en horarios inusuales: Identifica operaciones realizadas entre las 23:00 y 4:00



Para cada alerta, el sistema:

* Asigna un nivel de riesgo (Alto, Medio, Bajo)
* Registra la fecha y hora
* Proporciona una descripción detallada
* Mantiene referencia a la transacción original

```python
def detectar_fraude(transacciones):
    """
    Analiza transacciones bancarias para detectar posibles indicios de fraude.
    
    Parameters:
    transacciones (pandas.DataFrame): DataFrame con columnas:
        - fecha: datetime
        - monto: float
        - descripcion: str
        - lugar: str
        - tipo_transaccion: str
    
    Returns:
    pandas.DataFrame: Transacciones con indicadores de riesgo
    """
    # Crear copia para no modificar datos originales
    df = transacciones.copy()
    #df['fecha'] = pd.to_datetime(df['fecha'])
    
    # Inicializar columna de riesgo
    df['nivel_riesgo'] = 0
    df['alertas'] = ''
    
    # 1. Detectar transacciones de alto valor
    umbral_alto_valor = df['MOVIMIENTOS'].mean() + 2 * df['MOVIMIENTOS'].std()
    mask_alto_valor = df['MOVIMIENTOS'] > umbral_alto_valor
    df.loc[mask_alto_valor, 'nivel_riesgo'] += 2
    df.loc[mask_alto_valor, 'alertas'] += 'Monto inusualmente alto; '
    
    # 2. Detectar múltiples transacciones en corto tiempo
    def encontrar_transacciones_rapidas(grupo):
        if len(grupo) >= 3:  # Más de 3 transacciones
            tiempo_entre_trans = grupo['dateTimeStr'].diff()
            if (tiempo_entre_trans < timedelta(minutes=30)).sum() >= 2:
                return True
        return False
    
    df['fecha_hora'] = df['dateTimeStr'].dt.floor('h')
    grupos_tiempo = df.groupby(['fecha_hora'])
    transacciones_rapidas = grupos_tiempo.filter(encontrar_transacciones_rapidas)
    df.loc[df.index.isin(transacciones_rapidas.index), 'nivel_riesgo'] += 2
    df.loc[df.index.isin(transacciones_rapidas.index), 'alertas'] += 'Múltiples transacciones en corto tiempo; '
    
    # 3. Detectar patrones de "estructuración" (montos similares repetidos)
    montos_frecuentes = df.groupby('MOVIMIENTOS').size()
    montos_sospechosos = montos_frecuentes[montos_frecuentes >= 3].index
    mask_estructuracion = df['MOVIMIENTOS'].isin(montos_sospechosos)
    df.loc[mask_estructuracion, 'nivel_riesgo'] += 1
    df.loc[mask_estructuracion, 'alertas'] += 'Patrón de montos repetidos; '
    
    # 4. Detectar transacciones en horarios inusuales
    hora = df['dateTimeStr'].dt.hour
    mask_horario_inusual = (hora >= 23) | (hora <= 4)
    df.loc[mask_horario_inusual, 'nivel_riesgo'] += 1
    df.loc[mask_horario_inusual, 'alertas'] += 'Horario inusual; '
    
    # 5. Detectar cambios bruscos en patrones de gasto
    df['monto_promedio_7dias'] = df.sort_values('dateTimeStr').groupby(df['dateTimeStr'].dt.date)['MOVIMIENTOS'].transform('mean')
    df['variacion'] = abs(df['MOVIMIENTOS'] - df['monto_promedio_7dias']) / df['monto_promedio_7dias']
    mask_variacion = df['variacion'] > 3  # Variación mayor al 300%
    df.loc[mask_variacion, 'nivel_riesgo'] += 1
    df.loc[mask_variacion, 'alertas'] += 'Cambio brusco en patrón de gasto; '
    
    # Clasificar nivel de riesgo
    df['nivel_riesgo_categoria'] = pd.cut(df['nivel_riesgo'], 
                                        bins=[-float('inf'), 1, 2, float('inf')],
                                        labels=['Bajo', 'Medio', 'Alto'])
    
    # Limpiar alertas vacías
    df['alertas'] = df['alertas'].str.rstrip('; ')

    return df
    #return df[['fecha', 'monto', 'descripcion', 'lugar', 'tipo_transaccion', 
    #           'nivel_riesgo', 'nivel_riesgo_categoria', 'alertas']]
```

```python
import pandas as pd
import numpy as np

# Simulamos un registro de movimientos bancarios
data = {
    'fecha': ['2023-10-01', '2023-10-01', '2023-10-02', '2023-10-03', '2023-10-04', '2023-10-05'],
    'monto': [100, 200, 300, 12000, 150, 50000],  # Montos variados, incluyendo algunos altos
    'tipo': ['depósito', 'retiro', 'depósito', 'retiro', 'retiro', 'retiro'],
    'cuenta': ['123456', '123456', '123456', '123456', '123456', '123456'],
    'ubicacion': ['Madrid', 'Madrid', 'Madrid', 'Barcelona', 'Madrid', 'Tokyo']  # Ubicaciones variadas
}

# Convertimos los datos a un DataFrame
df = pd.DataFrame(data)

# Función para identificar transacciones sospechosas
def identificar_fraude(df):
    # Umbrales de seguridad
    umbral_monto_alto = 10000
    umbral_retiros_dia = 3  # Más de 3 retiros al día es sospechoso
    ubicaciones_sospechosas = ['Tokyo', 'Nueva York', 'Londres']  # Ubicaciones inusuales
    frecuencia_monto = df['monto'].mean() * 3  # Tres veces el promedio del monto

    # Identificar transacciones con montos inusuales
    transacciones_inusuales = df[df['monto'] > umbral_monto_alto]

    # Contar retiros por día
    df['fecha'] = pd.to_datetime(df['fecha'])
    retiros_por_dia = df[df['tipo'] == 'retiro'].groupby('fecha')['tipo'].count()
    retiros_sospechosos = retiros_por_dia[retiros_por_dia > umbral_retiros_dia]

    # Identificar transacciones desde ubicaciones sospechosas
    transacciones_ubicacion = df[df['ubicacion'].isin(ubicaciones_sospechosas)]

    # Identificar movimientos anómalos basados en el monto promedio
    transacciones_anomalas = df[(df['monto'] > frecuencia_monto) & (df['tipo'] == 'retiro')]

    # Combinar resultados
    resultados = pd.concat([transacciones_inusuales, transacciones_ubicacion, transacciones_anomalas]).drop_duplicates()

    # Agregar retiros sospechosos a los resultados
    if not retiros_sospechosos.empty:
        retiros_sospechosos_df = df[df['fecha'].isin(retiros_sospechosos.index) & (df['tipo'] == 'retiro')]
        resultados = pd.concat([resultados, retiros_sospechosos_df]).drop_duplicates()

    return resultados

# Ejecutar la función
resultados_fraude = identificar_fraude(df)

# Mostrar resultados
if not resultados_fraude.empty:
    print("Transacciones sospechosas encontradas:")
    print(resultados_fraude)
else:
    print("No se encontraron transacciones sospechosas.")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def detectar_fraude(transacciones):
    """
    Analiza transacciones bancarias para detectar posibles indicios de fraude.
    
    Parameters:
    transacciones (pandas.DataFrame): DataFrame con columnas:
        - fecha: datetime
        - monto: float
        - descripcion: str
        - lugar: str
        - tipo_transaccion: str
    
    Returns:
    pandas.DataFrame: Transacciones con indicadores de riesgo
    """
    # Crear copia para no modificar datos originales
    df = transacciones.copy()
    df['fecha'] = pd.to_datetime(df['fecha'])
    
    # Inicializar columna de riesgo
    df['nivel_riesgo'] = 0
    df['alertas'] = ''
    
    # 1. Detectar transacciones de alto valor
    umbral_alto_valor = df['monto'].mean() + 2 * df['monto'].std()
    mask_alto_valor = df['monto'] > umbral_alto_valor
    df.loc[mask_alto_valor, 'nivel_riesgo'] += 2
    df.loc[mask_alto_valor, 'alertas'] += 'Monto inusualmente alto; '
    
    # 2. Detectar múltiples transacciones en corto tiempo
    def encontrar_transacciones_rapidas(grupo):
        if len(grupo) >= 3:  # Más de 3 transacciones
            tiempo_entre_trans = grupo['fecha'].diff()
            if (tiempo_entre_trans < timedelta(minutes=30)).sum() >= 2:
                return True
        return False
    
    df['fecha_hora'] = df['fecha'].dt.floor('h')
    grupos_tiempo = df.groupby(['fecha_hora'])
    transacciones_rapidas = grupos_tiempo.filter(encontrar_transacciones_rapidas)
    df.loc[df.index.isin(transacciones_rapidas.index), 'nivel_riesgo'] += 2
    df.loc[df.index.isin(transacciones_rapidas.index), 'alertas'] += 'Múltiples transacciones en corto tiempo; '
    
    # 3. Detectar patrones de "estructuración" (montos similares repetidos)
    montos_frecuentes = df.groupby('monto').size()
    montos_sospechosos = montos_frecuentes[montos_frecuentes >= 3].index
    mask_estructuracion = df['monto'].isin(montos_sospechosos)
    df.loc[mask_estructuracion, 'nivel_riesgo'] += 1
    df.loc[mask_estructuracion, 'alertas'] += 'Patrón de montos repetidos; '
    
    # 4. Detectar transacciones en horarios inusuales
    hora = df['fecha'].dt.hour
    mask_horario_inusual = (hora >= 23) | (hora <= 4)
    df.loc[mask_horario_inusual, 'nivel_riesgo'] += 1
    df.loc[mask_horario_inusual, 'alertas'] += 'Horario inusual; '
    
    # 5. Detectar cambios bruscos en patrones de gasto
    df['monto_promedio_7dias'] = df.sort_values('fecha').groupby(df['fecha'].dt.date)['monto'].transform('mean')
    df['variacion'] = abs(df['monto'] - df['monto_promedio_7dias']) / df['monto_promedio_7dias']
    mask_variacion = df['variacion'] > 3  # Variación mayor al 300%
    df.loc[mask_variacion, 'nivel_riesgo'] += 1
    df.loc[mask_variacion, 'alertas'] += 'Cambio brusco en patrón de gasto; '
    
    # Clasificar nivel de riesgo
    df['nivel_riesgo_categoria'] = pd.cut(df['nivel_riesgo'], 
                                        bins=[-float('inf'), 1, 2, float('inf')],
                                        labels=['Bajo', 'Medio', 'Alto'])
    
    # Limpiar alertas vacías
    df['alertas'] = df['alertas'].str.rstrip('; ')
    
    return df[['fecha', 'monto', 'descripcion', 'lugar', 'tipo_transaccion', 
               'nivel_riesgo', 'nivel_riesgo_categoria', 'alertas']]

# Ejemplo de uso
def crear_ejemplo():
    """Crea un DataFrame de ejemplo con transacciones"""
    data = {
        'fecha': pd.date_range(start='2024-01-01', periods=100, freq='h'),
        'monto': np.random.normal(1000, 500, 100),
        'descripcion': ['Compra'] * 100,
        'lugar': ['Tienda'] * 100,
        'tipo_transaccion': ['Débito'] * 100
    }
    
    # Agregar algunas transacciones sospechosas
    data['monto'][10:13] = 5000  # Montos altos
    data['monto'][20:23] = 999.99  # Estructuración
    
    return pd.DataFrame(data)
```
